## Policzenie wynikowych entropii prawdopodobieństw modeli w przykładach

```python
df['SM_prob_np'] = df['SM_prob'].apply(lambda x: np.array(ast.literal_eval(x)))
df['SMA_prob_np'] = df['SMA_prob'].apply(lambda x: np.array(ast.literal_eval(x)))
```

```python
df['SM_prob_entropy'] = df['SM_prob_np'].apply(lambda x: -(x*np.log2(x)).sum())
df['SMA_prob_entropy'] = df['SMA_prob_np'].apply(lambda x: -(x*np.log2(x)).sum())
```
