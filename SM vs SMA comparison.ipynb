```python
import pandas as pd
```

```python
df = pd.read_csv("sm_vs_sma_dataset2.csv").drop(['Unnamed: 0'], axis=1)
```

## Co powoduje, że SM był poprawny a SMA nie, gdzie oba modele były pewne swojej decyzji, ale SM trochę bardziej?

```python
SM_correct_SMA_incorrect_SM_dominance = df[(df['SM_pred'] == df['label_num']) & (df['SMA_pred'] != df['label_num'])].\
    sort_values(['SM_highest_prob', 'SMA_highest_prob'], ascending=False).\
    head(10)
```

```python
SM_correct_SMA_incorrect_SM_dominance[['SM_pred', 'SMA_pred', 'label_num', 'SM_highest_prob', 'SMA_highest_prob']].head(2)
```
