```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
def get_metrics(df, train_name, val_name):
    df2 = {}
    for i, el in df.iterrows():
        if el['step'] not in df2:
            df2[el['step']] = {}
        df2[el['step']][el['metric']] = el['value']

    df3 = {
        'step': [],
        'Zbi贸r treningowy': [],
        'Zbi贸r walidacyjny': []
    }
    for i, x in df2.items():
        df3['step'].append(i)
        df3['Zbi贸r treningowy'].append(df2[i][train_name])
        df3['Zbi贸r walidacyjny'].append(df2[i][val_name])
    return pd.DataFrame.from_dict(df3)
```

```python
df = pd.read_csv('finetuned/train_acc-val_acc.csv')
ax = get_metrics(df, 'train_acc', 'val_acc').plot(x='step',
                                            y=['Zbi贸r treningowy', 'Zbi贸r walidacyjny'])
ax.set_xlabel('Epoka')
ax.set_ylabel('Dokadno')
#ax.set_title('Por贸wnanie dokadnoci modelu\nLiarPlusSingleFinetunedRoBERTaClassifier')

plt.axvline(x=17, color='grey', linestyle='--')
plt.axhline(y=0.338785046728972, color='grey', linestyle='--')

ticks = list(range(0, 31, 5)) + [17]
ticks.sort()

ax.set_xticks(ticks)

def fmt_tick(t):
    if type(t) is float or isinstance(t, float):
        return f'{round(t, 3):.3f}'
    return str(t)

ax.set_xticklabels([fmt_tick(t) for t in ticks])

ticks = list(ax.get_yticks())

if 0.339 not in ticks:
    ticks.append(0.339)
    ticks.sort()

ax.set_yticks(ticks)

ax.set_yticklabels([fmt_tick(t) for t in ticks])

plt.savefig('finetuned/finetuned_acc_comparison.png', bbox_inches='tight')
plt.show()
```
