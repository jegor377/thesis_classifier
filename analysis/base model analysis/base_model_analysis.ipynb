```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
df = pd.read_csv('train_acc-val_acc.csv')
```

```python
df.head()
```

```python
df2 = {}
for i, el in df.iterrows():
    if el['step'] not in df2:
        df2[el['step']] = {}
    df2[el['step']][el['metric']] = el['value']
```

```python
df3 = {
    'step': [],
    'Zbiór treningowy': [],
    'Zbiór walidacyjny': []
}
for i, x in df2.items():
    df3['step'].append(i)
    df3['Zbiór treningowy'].append(df2[i]['train_acc'])
    df3['Zbiór walidacyjny'].append(df2[i]['val_acc'])
df3 = pd.DataFrame.from_dict(df3)
```

```python
df3
```

```python
ax = df3.plot(x='step', y=['Zbiór treningowy', 'Zbiór walidacyjny'])
#ax.set_title('Porównanie dokładności na zbiorze\ntreningowym i walidacyjnym podczas trenowania')
ax.set_xlabel('Epoka')
ax.set_ylabel('Dokładność')

plt.savefig('base_model_train_vs_val_acc.png', bbox_inches='tight')
plt.show()
```
