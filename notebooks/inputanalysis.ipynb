```python
import matplotlib.pyplot as plt

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

test_df = pd.read_csv("/kaggle/input/articles/test2.csv")
train_df = pd.read_csv("/kaggle/input/articles/train2.csv")
val_df = pd.read_csv("/kaggle/input/articles/val2.csv")
```

```python
test_df = test_df[[
    "subject", "speaker", "job_title", "state",
    "party_affiliation", "context", "sentiment", "question",
    "curse", "emotion", "gibberish", "offensiveness", "political_bias"
]]

train_df = train_df[[
    "subject", "speaker", "job_title", "state",
    "party_affiliation", "context", "sentiment", "question",
    "curse", "emotion", "gibberish", "offensiveness", "political_bias"
]]

val_df = val_df[[
    "subject", "speaker", "job_title", "state",
    "party_affiliation", "context", "sentiment", "question",
    "curse", "emotion", "gibberish", "offensiveness", "political_bias"
]]
```

```python
for col in test_df.columns:
    if test_df[col].dtype == object or pd.api.types.is_string_dtype(test_df[col]):
        max_len = test_df[col].astype(str).str.len().max()
        print(f"{col}: {max_len}")

for col in test_df.columns:
    if test_df[col].dtype == object or pd.api.types.is_string_dtype(test_df[col]):
        longest_value = test_df.loc[test_df[col].astype(str).str.len().idxmax(), col]
        print(f"{col}: {longest_value}")
        
for col in test_df.columns:
    if col in ["political_bias", "emotion", "curse", "gibberish", "offensiveness", "question", "sentiment"]:
        continue
    if test_df[col].dtype == object or pd.api.types.is_string_dtype(test_df[col]):
        lengths = test_df[col].astype(str).str.len()
        distribution = lengths.value_counts().sort_index()
        
        plt.figure(figsize=(10, 4))
        plt.bar(distribution.index, distribution.values, color='skyblue')
        plt.title(f'Rozkład długości tekstów w kolumnie: {col}')
        plt.xlabel('Liczba znaków')
        plt.ylabel('Liczba wartości')
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.tight_layout()
        plt.show()
```
