```python
import pandas as pd
import numpy as np
import ast
from matplotlib import pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
plt.style.use("ggplot")
```

```python
sm_correct_sma_incorrect = pd.read_csv('analysis/SM_correct_SMA_incorrect_csv.csv')
```

```python
sm_correct_sma_incorrect_stats = sm_correct_sma_incorrect.drop('id', axis=1).\
    describe().\
    transpose().\
    drop('count', axis=1).\
    round(decimals=3)
sm_correct_sma_incorrect_stats
```

```python
sm_correct_sma_incorrect_stats.to_csv('analysis/results/sm_correct_sma_incorrect_stats.csv')
```

```python
sm_incorrect_sma_correct = pd.read_csv('analysis/SM_incorrect_SMA_correct_csv.csv')
```

```python
sm_incorrect_sma_correct_stats = sm_incorrect_sma_correct.drop('id', axis=1).\
    describe().\
    transpose().\
    drop('count', axis=1).\
    round(decimals=3)
sm_incorrect_sma_correct_stats
```

```python
sm_incorrect_sma_correct_stats.to_csv('analysis/results/sm_incorrect_sma_correct_stats.csv')
```

# Analiza sm_correct_sma_incorrect

```python
vc = sm_correct_sma_incorrect['label_num'].value_counts().sort_index()
ax = vc.\
    plot(kind='barh', title='Rozkład wartości label_num w sm_correct_sma_incorrect')
ax.set_xlabel('Liczba przykładów')

# Dodanie wartości na końcu słupków
for i, (label, count) in enumerate(vc.items()):
    ax.text(count - (1.5 if count > 10 else 1), i, str(count), va='center', c='w')

plt.tight_layout()
plt.savefig('analysis/results/sm_correct_sma_incorrect__label_num_distribution.png')
plt.show()
```

```python
vc = sm_correct_sma_incorrect['SMA_pred'].value_counts().sort_index()
ax = vc.\
    plot(kind='barh', title='Rozkład wartości SMA_pred w sm_correct_sma_incorrect')
ax.set_xlabel('Liczba przykładów')

# Dodanie wartości na końcu słupków
for i, (label, count) in enumerate(vc.items()):
    ax.text(count - (0.75 if count > 10 else 0.5), i, str(count), va='center', c='w')

plt.tight_layout()
plt.savefig('analysis/results/sm_correct_sma_incorrect__SMA_pred_distribution.png')
plt.show()
```
