```python
import pandas as pd
import numpy as np
import ast
```

```python
sm_correct_sma_incorrect = pd.read_csv('analysis/SM_correct_SMA_incorrect_csv.csv')
```

```python
sm_correct_sma_incorrect['SM_prob_np'] = sm_correct_sma_incorrect['SM_prob'].apply(lambda x: np.array(ast.literal_eval(x)))
sm_correct_sma_incorrect['SMA_prob_np'] = sm_correct_sma_incorrect['SMA_prob'].apply(lambda x: np.array(ast.literal_eval(x)))
```

```python
sm_correct_sma_incorrect['SM_prob_entropy'] = sm_correct_sma_incorrect['SM_prob_np'].apply(lambda x: -(x*np.log2(x)).sum())
sm_correct_sma_incorrect['SMA_prob_entropy'] = sm_correct_sma_incorrect['SMA_prob_np'].apply(lambda x: -(x*np.log2(x)).sum())
```
